# 파이프라인 구축

```python
X = train.drop("비만유형", axis=1)
y = train["비만유형"]
```

```python
X
```

```python

# Preprocessing for numerical data:
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_columns.remove('비만유형')
# Preprocessing for categorical data: one-hot encoding
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Create the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)])
```

```python
# Fit the preprocessor
preprocessor.fit(X)

# Transform train data
X_train = preprocessor.transform(X)

# Transform test data
id = test['id']
test = preprocessor.transform(test)
```

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y)
```

# 모델링

```python
'''

Optuna는 머신 러닝 알고리즘의 최적 하이퍼파라미터를 
자동으로 찾기 위해 설계된 오픈 소스 하이퍼파라미터 최적화 프레임워크
이는 다양한 최적화 알고리즘을 지원하며, 복잡한 모델의 최적 매개변수를 효과적으로 찾는 데 
사용되며 학계와 산업 모두에서 널리 사용







'''
```

```python
def objective(trial):
    
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 1200, 1500),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.02),
        'gamma': trial.suggest_loguniform('gamma', 1e-2, 1.0),
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-2, 0.1),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-2, 0.1),
        'max_depth': trial.suggest_int('max_depth', 4, 6),
        'min_child_weight': trial.suggest_int('min_child_weight', 4, 6),
        'subsample': trial.suggest_uniform('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.5, 1.0),
    }

    # Initialize and train the model
    model = XGBClassifier(**param, random_state=42, use_label_encoder=False, eval_metric='logloss')

    # Perform cross-validation
    score = cross_val_score(model, np.array(X_train), y_train, scoring='accuracy', cv=5, n_jobs=-1).mean()
    print("The Accuracy: ", score)
    
    model.fit(np.array(X_train), y_train)
    
    return score

# Enable Optuna's progress bar
optuna.logging.set_verbosity(optuna.logging.INFO)

study = optuna.create_study(direction='maximize')
    
# Start the optimization process
study.optimize(objective, n_trials=100, show_progress_bar=True)




best_trial = study.best_trial

print("Best trial:")
print(f"  Value (Accuracy): {best_trial.value}")

print("Best hyperparameters:")
for key, value in best_trial.params.items():
    print(f"  {key}: {value}")


from optuna.visualization import plot_optimization_history, plot_param_importances

# Plot optimization history
plot_optimization_history(study)

# Plot parameter importances
plot_param_importances(study)
```

```python
params = {
    'n_estimators': 1312,
    'learning_rate': 0.018279520260162645,
    'gamma': 0.0024196354156454324,
    'reg_alpha': 0.9025931173755949,
    'reg_lambda': 0.06835667255875388,
    'max_depth': 5,
    'min_child_weight': 5,
    'subsample': 0.883274050086088,
    'colsample_bytree': 0.6579828557036317
}


# **params -> 키와 값을 해당하는 인수로 전달해줌
model = XGBClassifier(random_state=42, **params)

score = cross_val_score(model, np.array(X_train), y_train, scoring='accuracy', cv=5, n_jobs=-1).mean()
print("Accuracy: ", score)
model.fit(np.array(X_train), y_train)
```
